Bajar el CSV de Oferta GastronÃ³mica desde Buenos Aires Data (https://data.buenosaires.gob.ar/dataset/). Idealmente hacer esto con Python.

```python
import pandas as pd
import requests
import io

bajada = requests.get('https://cdn.buenosaires.gob.ar/datosabiertos/datasets/ente-de-turismo/oferta-establecimientos-gastronomicos/oferta_gastronomica.csv')
data = bajada.content.decode('utf-8')
df = pd.read_csv(io.StringIO(data), delimiter=',', decimal='.', encoding= 'utf-8') # El StringIO me trata la cadena como un archivo

# Los meto a una lista de tuplas para que los pueda mandar a la BD
entradas = [  ( str(row['nombre']), str(row['categoria']), str(row['direccion_completa']), str(row['barrio']), str(row['comuna'][7::]) ) for i, row in df.iterrows()]
```
